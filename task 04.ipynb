```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r'C:\twitter_training.csv'
df = pd.read_csv(file_path)
```

```python
# Renaming the columns
df.columns = ['id', 'entity', 'sentiment', 'text']
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
sentiment_counts = df['sentiment'].value_counts()
sentiment_percentage = sentiment_counts / len(df) * 100
print(sentiment_percentage)
```

```python
# Sentiment distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sentiment', hue='sentiment', palette='viridis', legend=False)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Entity-wise sentiment distribution
plt.figure(figsize=(10, 6)) 
ax = sns.countplot(data=df, x='entity', hue='sentiment', palette='viridis')
plt.title('Entity-wise Sentiment Distribution')
plt.xlabel('Entity')
plt.ylabel('Entity Count')
plt.xticks(rotation=90)  
plt.legend(title='Sentiment')
for label in ax.get_xticklabels():
    label.set_fontsize(8)
plt.tight_layout() 
plt.show()
```
